Represente a função de transferência do controlador P em que o ganho proporcional $k_p = 1$. A função de transferência do processo é $H(s) = \frac{3}{s^2 +0.4s+4}$. Determine a função de transferência da referência para a saída $\frac{Y(s)}{R(s)}$, da referência para o erro $\frac{E(s)}{R(s)}$, e da referência para o sinal de controle $\frac{U(s)}{R(s)}$. Represente o sistema de controle em malha fechada e o simule para uma entrada de referência igual ao degrau unitário. Apresente os sinais da saída $y(t)$, erro $e(t)$, e da entrada $u(t)$ no tempo.

```python
from control.matlab import *
import matplotlib.pyplot as plt

kp = 1

s = tf('s')

G = kp
print(G)

H = 3/(s**2+0.4*s+4)
print(H)

pzmap(H)
plt.show()

RY = feedback(G*H, 1, sign =- 1) # FT malha fechada, realimentação negativa
print(RY)
pzmap(RY)
plt.show()

RE = feedback(1, G*H, sign =- 1)
print(RE)

RU = feedback(G, H, sign =- 1)
print(RU)

#Instante de tempo final das simulações
T=50

y,t = step(RY, T)
plt.figure(1)
plt.plot(t,y) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Saída, y')
plt.show()

e,t = step(RE, T)
plt.figure(2)
plt.plot(t,e) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Erro, e')
plt.show()

u,t = step(RU, T)
plt.figure(3)
plt.plot(t,u) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Entrada, u')
plt.show()
```

```python
#Um sistema na forma padrão de segunda ordem
from control.matlab import *
import matplotlib.pyplot as plt

s = tf('s')

wn = 3.0
zeta = 0.2
H=wn**2/(s**2 + 2*zeta*wn*s + wn**2)
print(H)

y,t = step(H)
plt.figure(1)
plt.plot(t,y) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao degrau em malha aberta, theta')
plt.show()

S = stepinfo(H, SettlingTimeThreshold=0.01, RiseTimeLimits=(0.0, 1.0))
print(S)
print('Tempo de subida:', S["RiseTime"])
print('Tempo de pico:', S["PeakTime"])
print('Pico:', S["Peak"])
print('Sobre-sinal (%):', S["Overshoot"])
print('Tempo de acomodação:', S["SettlingTime"])
print('Valor de regime permanente:', S["SteadyStateValue"])
```
