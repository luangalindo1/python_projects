# Exercício

Determine a função de transferência de $R(s)$ para $Y(s)$, ou seja $\frac{Y(s)}{R(s)}$, do diagrama de blocos da figura a seguir, em que $G_1(s) = \frac{1}{s^2}$ e $G_2(s) = \frac{1}{10s+1}$.

Captura de Tela 2021-11-28 às 16.37.31.png

```python
#Resposta

from control.matlab import *

s = tf('s')
G1 = 1/s**2
G2 = 1/(10*s + 1)

H_r0 = 1
H_r1 = feedback(G1, H_r0, sign = -1)

H_mf = parallel(H_r1, G2)

print(H_mf)
```
