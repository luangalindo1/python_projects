Vamos representar um sistema por meio de sua função de transferência $H(s) = \frac{1}{s+2}$ e depois simulá-lo para diferentes entradas

```python
#Representação de um sistema por meio de sua função de transferência
import numpy as np
from control.matlab import *
import matplotlib.pyplot as plt

H=tf (1 ,[1 ,2])
print(H)

y,t = impulse(H)
plt.figure(1)
plt.plot(t,y) 
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao impulso, y')
plt.grid(True)
plt.show()

y,t = step(H)
plt.figure(2)
plt.plot(t,y) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao degrau, y')
plt.show()

t = np.linspace(0, 3.5, num = 500)
u = np.sin(10*t)
y,t,x = lsim(H, u, t)
plt.figure(3)
plt.plot(t,y) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao sinal sen(10t), y')
plt.show()

plt.figure(4)
plt.plot(t,u) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Entrada, u = sen(10t)')
plt.show()
```

# Exemplo

Vamos simular um sistema representado pela equação diferencial $\dot{y} + 2y = u$. A entrada $u(t) = \mathbf{1} (t)$, e a condição inicial é $y(0) = 0$.

```python
#Representação e resolução de uma equação diferencial
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Função que retorna doty = dy/dt
def modelo(y,t):
    u = 1
    doty = -2 * y + u
    return doty

# Condição inicial
y0 = 0

# Vetor de tempo
t = np.linspace(0,3.5)

# Resolver a equação diferencial ordinária
y = odeint(modelo,y0,t)

# Apresentar os resultados
plt.plot(t,y)
plt.xlabel('Tempo, t')
plt.ylabel('Resposta ao degrau, y')
plt.show()
```

Representação e simulação de um sistema de 2ª ordem por meio de variáveis de estado.

```python
#Representação de um sistema de 2ª ordem
#por meio de variáveis de estado
from control.matlab import *
import numpy as np
import matplotlib.pyplot as plt

A = np.matrix([[ 0,  1],
               [-4, -2]])
B = np.matrix([[0],
               [1]])
C = np.matrix([[1, 0]])
d = np.matrix([0])
Hss=ss(A, B, C, d)
print(Hss)

y,t,x = impulse(Hss, return_x=True)
plt.figure(1)
plt.plot(t,y) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao impulso, y')
plt.show()

plt.figure(2)
plt.plot(t,x) 
plt.grid(True)
plt.legend(['x1','x2'])
plt.xlabel('Tempo, t') 
plt.ylabel('Estados para a resposta ao impulso, x')
plt.show()

y,t,x = step(Hss, return_x=True)
plt.figure(3)
plt.plot(t,y) 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao degrau, y')
plt.show()

plt.figure(4)
plt.plot(t,x) 
plt.grid(True)
plt.legend(['x1','x2'])
plt.xlabel('Tempo, t') 
plt.ylabel('Estados para a resposta ao degrau, x')
plt.show()
```

```python
#Obtenção dos polos de um sistema representado por meio de variáveis de estado
from control.matlab import *
import numpy as np
import matplotlib.pyplot as plt

A = np.matrix([[ 0,  1],
               [-4, -2]])
B = np.matrix([[0],
               [1]])
C = np.matrix([[1, 0]])
d = np.matrix([0])
Hss=ss(A, B, C, d)
print(Hss)

H = ss2tf(Hss)

print('Polos do sistema:')
p = pole(H)
print(p)

print('\nAutovalores da matriz A:')
autovaloresEautovetores = np.linalg.eig(A)
print(autovaloresEautovetores[0])
```
