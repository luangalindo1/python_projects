# Exemplos

Vamos definir matrizes

```
from numpy import *

#Definir a matriz
print ('A = ')
A = matrix([[1, 2],
            [3, 4]])

print(A)

# Identificar as dimensões da matriz
dimensoes_matriz_A = A.shape
print ('\n')
print('Dimensões da matriz: ', dimensoes_matriz_A)

# Acessar um elemento da matriz
elemento = A[0, 1]
print ('\n')
print ('A(1, 2) =  ', elemento)

# Acessar uma linha da matriz
linha = A[0, :]
print ('\n')
print ('A(1, :) =  ', linha)

# Acessar uma coluna da matriz
coluna = A[:, 0]
print ('\n')
print ('A(:, 1) =  ')
print(coluna)
```

Vamos criar matrizes de zeros, de uns, de uma constante, e a matriz identidade.

```
# Criar uma matriz com todos os elementos iguais a zero
B = zeros((3, 3))

print ('Matriz 3 x 3 com todos os elementos iguais a zero:')
print (B)

# Criar uma matriz com todos os elementos iguais a um.
C = ones((2, 2))

print ('\n')
print ('Matriz 2 x 2 com todos os elementos iguais a um:')
print (C)

# Criar uma matriz com todos os elementos iguais a uma constante
D = 7.8 * (ones((3, 2)))

print ('\n')
print ('Matriz 3 x 2 com todos os elementos iguais a uma constante:')
print (D)

# Cria uma matriz identidade 5 x 5
I55 = identity(5)

print ('\n')
print ('Matriz identidade 5 x 5')
print (I55)
```

Operações matriciais

```
from numpy import *

#Definir a matriz
print ('A=')
A = matrix([[1, 2],
            [3, 4]])

print(A)

# Determinar a transposta da matriz
A_t = transpose(A)

print ('\n')
print ('A^T = ')
print (A_t)

# Determinar a inversa da matriz
A_inv = linalg.inv(A)

print ('\n')
print ('A^(-1) = ')
print (A_inv)

# Determinante da matriz
detA = linalg.det(A)

print ('\n')
print ('det(A)=')
print (detA)

# Autovalores e autovetores da matriz
eigA = linalg.eig(A)

print ('\n')
print ('Autovalores da matriz:')
print (eigA[0])

print ('\n')
print ('Autovetores da matriz:')
print (eigA[1])
```

Operações matriciais

```
from numpy import *

#Definir a matriz
print('A=')
A = matrix([[1, 2],
            [3, 4]])

print(A)

#Multiplicar a matriz por uma constante
B = 10.0*A
print('B=')

print(B)

# Soma das matrizes
C = A + B

print('\n')
print('Soma das matrizes:')
print(C)

# Subtração das matrizes
D = A - B

print('\n')
print('Subtração das matrizes:')
print(D)

# Potência
A2 = linalg.matrix_power(A, 2)
print('\n')
print('Matriz ao quadrado:')
print(A2)
```

Vamos concatenar vetores e matrizes

```
from numpy import *

a = matrix([[1],
            [3]])
print('Vetor a =')
print(a)

b = matrix([[2],
            [4]])
print('\n')
print('Vetor b =')
print(b)

C = concatenate((a, b), axis=1)
print('\n')
print('Matriz C =')
print(C)

d = concatenate((a, b), axis=0)
print('\n')
print('Vetor d =')
print(d)
```

# Exemplo 2

Vamos simular um sistema de controle com realimentação de estados e um observador de estados.

```
from control import *
import matplotlib.pyplot as plt
from numpy import *
from control.matlab import ss

A = matrix([[-7.9630,   43.4815,   12.4444,   46.3704],
     [ 1.2206,    1.6071,   -7.3527,   -5.8808],
     [-0.0395,    7.0958,   -4.9734,    0.9750],
     [-1.2214,    1.2520,    5.8430,    6.3293]])

B = matrix([[23],
     [-1],
     [ 2],
     [ 2]])

C = matrix([[-0.0330,    0.2230,    0.1039,    0.3873]])
     
D = matrix([0])

processo = ss(A, B, C, D)

K = matrix([[23.7713,  318.9839, -195.7440,   90.3655]])

N = 840.0000

#Representação do sistema de controle em malha fechada sem observador
MF1 = ss(A-B*K,N*B,C,0)

t1,y1 = step_response(MF1)

L = matrix([[-4420],
            [ -210],
            [-1925],
            [ 325]])

Al1 = concatenate((A-B*K, B*K), axis=1)
Al2 = concatenate((zeros(A.shape), A-L*C), axis=1)
AMF = concatenate((Al1, Al2), axis=0)

BMF = concatenate((N*B, zeros(B.shape)), axis=0)

CMF = concatenate((C, zeros(C.shape)), axis=1)

MF2 = ss(AMF,BMF,CMF,0)

t2,y2 = step_response(MF2,X0=concatenate((zeros(B.shape), 0.1*ones(B.shape)), axis=0))
plt.figure(1)
plt.plot(t1,y1,'r',t2,y2,'g') 
plt.xlabel('Tempo, t') 
plt.ylabel("Resposta ao degrau em malha fechada, y")
plt.legend(["Sem observador","Com observador"])
plt.show()
```
